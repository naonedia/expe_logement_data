```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(
    'data/finance.csv',
        delimiter=',',
        header=0,
       index_col=False
)
df.head()
```

```python
df['annee'] = df['annee'].astype('int64')
df['trimestre'] = df['trimestre'].apply(lambda t: int(t[1]))
df.head()
```

```python
df.sort_values(['annee', 'trimestre'], ascending=[False, False], inplace=True)
df.head()
```

```python
df.drop(df[df['annee'] < 2000].index, inplace=True)
df.head()
```

```python
df.drop('trimestre', axis=1, inplace=True)
df['rdb'] = df['rdb'].astype('float64') / 100 + 1
df['rdb_uc'] = df['rdb_uc'].astype('float64') / 100 + 1
df['pa_rdb'] = df['pa_rdb'].astype('float64') / 100 + 1
df['pa_rdb_uc'] = df['pa_rdb_uc'].astype('float64') / 100 + 1
df['taux_epargne'] = df['taux_epargne'].astype('float64') / 100 + 1
df['taux_epargne_financier'] = df['taux_epargne_financier'].astype('float64') / 100 + 1
df = df.groupby('annee', as_index=False).prod()
df
```

```python
df.loc[0, 'rdb_c'] = 100
df.loc[0, 'rdb_uc_c'] = 100
df.loc[0, 'pa_rdb_c'] = 100
df.loc[0, 'pa_rdb_uc_c'] = 100
df.loc[0, 'taux_epargne_c'] = 100
df.loc[0, 'taux_epargne_financier_c'] = 100

for i in range(1, len(df)):
    df.loc[i, 'rdb_c'] = df.loc[i-1, 'rdb_c'] * df.loc[i, 'rdb']
    df.loc[i, 'rdb_uc_c'] = df.loc[i-1, 'rdb_uc_c'] * df.loc[i, 'rdb_uc']
    df.loc[i, 'pa_rdb_c'] = df.loc[i-1, 'pa_rdb_c'] * df.loc[i, 'pa_rdb']
    df.loc[i, 'pa_rdb_uc_c'] = df.loc[i-1, 'pa_rdb_uc_c'] * df.loc[i, 'pa_rdb_uc']
    df.loc[i, 'taux_epargne_c'] = df.loc[i-1, 'taux_epargne_c'] * df.loc[i, 'taux_epargne']
    df.loc[i, 'taux_epargne_financier_c'] = df.loc[i-1, 'taux_epargne_financier_c'] * df.loc[i, 'taux_epargne_financier']

df
```

```python
for i in range(5, len(df)):
    for j in range(1, 6):
        df.loc[i, 'rdb_m%i' % j] = df.loc[i-j, 'rdb']
        df.loc[i, 'rdb_c_m%i' % j] = df.loc[i-j, 'rdb_c']
        df.loc[i, 'rdb_uc_m%i' % j] = df.loc[i-j, 'rdb_uc']
        df.loc[i, 'rdb_uc_c_m%i' % j] = df.loc[i-j, 'rdb_uc_c']
        df.loc[i, 'pa_rdb_m%i' % j] = df.loc[i-j, 'pa_rdb']
        df.loc[i, 'pa_rdb_c_m%i' % j] = df.loc[i-j, 'pa_rdb_c']
        df.loc[i, 'pa_rdb_uc_m%i' % j] = df.loc[i-j, 'pa_rdb_uc']
        df.loc[i, 'pa_rdb_uc_c_m%i' % j] = df.loc[i-j, 'pa_rdb_uc_c']
        df.loc[i, 'taux_epargne_m%i' % j] = df.loc[i-j, 'taux_epargne']
        df.loc[i, 'taux_epargne_c_m%i' % j] = df.loc[i-j, 'taux_epargne_c']
        df.loc[i, 'taux_epargne_financier_m%i' % j] = df.loc[i-j, 'taux_epargne_financier']
        df.loc[i, 'taux_epargne_financier_c_m%i' % j] = df.loc[i-j, 'taux_epargne_financier_c']

df
```
