```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv(
        '../expe_logement_dataset/datasets/origin_enriched_v3.csv',
        header=0,
        index_col=False
    )
df.head()
```

```python
qiles = pd.qcut(x=df.valeur_fonciere, q=500, duplicates='drop')
counts = qiles.value_counts()
rc = counts.reindex(qiles.cat.categories)

fig = go.Figure(data=go.Scatter(x=list(map(lambda x: str(x), rc.index)), y=rc.values, fill='tozeroy'))
fig.show()
```

```python
qiles = pd.qcut(x=df.surface_reelle_bati, q=2500, duplicates='drop')
counts = qiles.value_counts()
rc = counts.reindex(qiles.cat.categories)

fig = go.Figure(data=go.Scatter(x=list(map(lambda x: str(x), rc.index)), y=rc.values, fill='tozeroy'))
fig.show()
```

```python
qiles = pd.qcut(x=(df.valeur_fonciere)/(df.surface_reelle_bati), q=2500, duplicates='drop')
counts = qiles.value_counts()
rc = counts.reindex(qiles.cat.categories)

fig = go.Figure(data=go.Scatter(x=list(map(lambda x: str(x), rc.index)), y=rc.values, fill='tozeroy'))
fig.show()
```

```python
df.drop(df[(df.valeur_fonciere < 45000) | (df.valeur_fonciere > 1000000)].index, inplace=True)
#df.drop(df[(df.surface_reelle_bati < 13)].index, inplace=True)
df.drop(df[((df.valeur_fonciere)/(df.surface_reelle_bati) < 1200) | ((df.valeur_fonciere)/(df.surface_reelle_bati) > 10000)].index, inplace=True)
df.head()
```

```python
#df.drop(['date_mutation', 'mois_sin', 'mois_cos', 'longitude', 'latitude'], axis=1, inplace=True)
df.drop(['date_mutation', 'mois_sin', 'mois_cos'], axis=1, inplace=True)

df.drop(['nom_commune'], axis=1, inplace=True)

df.head()
```

```python
for col in df.columns:
    if col.endswith('/zoo') or col.startswith('/zoo'):
        df.drop(col, axis=1, inplace=True)
df.head()
```

```python
df['valeur_fonciere'] = df['valeur_fonciere'].astype(np.int32)
df['code_postal'] = df['code_postal'].astype(str)
df['surface_carrez'] = df['surface_carrez'].astype(np.int16)
df['surface_reelle_bati'] = df['surface_reelle_bati'].astype(np.int16)
df['nombre_pieces_principales'] = df['nombre_pieces_principales'].astype(np.int8)
df['surface_terrain'] = df['surface_terrain'].astype(np.int16)
df['annee'] = df['annee'].astype(np.int16)
df['mois'] = df['mois'].astype(np.int8)
for col in df.columns:
    if col.startswith('foot/') or col.startswith('car/'):
        df[col] = df[col].astype(np.int16)
df.head()
```

```python
df['surface_non_bati'] = df['surface_terrain'] - df['surface_reelle_bati']
df.head()
```

```python
for y in range(0,5):
    for t in range (1,5):
        for v in ['rdb', 'rdb_uc', 'pa_rdb', 'pa_rdb_uc', 'taux_epargne', 'taux_epargne_financier']:
            if (y not in [0, 1]) or (t != 1):
                col = "%i-%i-%s" % (y, t, v)
                df.drop(col, axis=1, inplace=True)
df.head()
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
df.drop('type_local_Maison', axis=1, inplace=True)
df.head()
```

```python
df = df.T.drop_duplicates(keep='last').T
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Taille du dataset de test (fraction du dataset total)
TEST_SIZE = 0.25

# Random seed pour la sÃ©lection des observations
SPLIT_RANDOM_SEED = 42

# X et y
X = df.drop("valeur_fonciere", axis=1)
y = df['valeur_fonciere']

# Split du dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=SPLIT_RANDOM_SEED)
```

```python
print('Train dataset : X_train shape ' + str(X_train.shape) + ' et y_train shape ' + str(y_train.shape))
print('Test dataset : X_test shape ' + str(X_test.shape) + ' et y_test shape ' + str(y_test.shape))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, median_absolute_error

def meaae(y_true, y_pred):
    return mean_absolute_error(y_true, y_pred)

def medae(y_true, y_pred):
    return median_absolute_error(y_true, y_pred)

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import ExtraTreeRegressor

TRAIN_RANDOM_SEED=42

# Instantiation du modÃ¨le.
abr = AdaBoostRegressor(
    base_estimator=ExtraTreeRegressor(
        max_depth=32,
        criterion='mse',
        max_features=22
    ),
    n_estimators=50,
    random_state=TRAIN_RANDOM_SEED
)

# EntraÃ®nement du modÃ¨le
abr.fit(X_train, y_train)

# PrÃ©diction sur le corpus d'apprentissage afin de calculer l'erreur en prÃ©diction
y_train_predicted = abr.predict(X_train)
print("MeaAE en prÃ©diction : %i" % meaae(y_train, y_train_predicted))
print("MedAE en prÃ©diction : %i" % medae(y_train, y_train_predicted))
print("RMSE en prÃ©diction : %i" % rmse(y_train, y_train_predicted))

# PrÃ©diction sur le corpus de test afin de calculer l'erreur en gÃ©nÃ©ralisation
y_test_predicted = abr.predict(X_test)
print("MeaAE en gÃ©nÃ©ralisation : %i" % meaae(y_test, y_test_predicted))
print("MedAE en gÃ©nÃ©ralisation : %i" % medae(y_test, y_test_predicted))
print("RMSE en gÃ©nÃ©ralisation : %i" % rmse(y_test, y_test_predicted))
```

```python
s = 5

pe = ((y_test - y_test_predicted).abs() / y_test) * 100
rpe = np.floor(pe / s) * s

counts = rpe.value_counts()
counts = counts.reindex(index=np.sort(counts.index)).cumsum()

fig = go.Figure(data=go.Scatter(x=list(map(lambda x: "[0;%i[" % (x+s), counts.index)), y=(counts.values / len(rpe)) * 100, fill='tozeroy'))
fig.show()
```

```python
# Obtention de l'importance des features
importances = abr.feature_importances_

# CrÃ©ation d'un DataFrame pour associer chaque feature Ã  son importance
importances = pd.concat([pd.Series(X_train.columns), pd.Series(importances * 100)], axis=1).rename({0: 'feature', 1: 'importance'}, axis=1)

# Tri
importances.sort_values(by='importance', ascending=False, inplace=True)

#Affichage
print(importances.to_string())
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import ExtraTreeRegressor

TRAIN_RANDOM_SEED=42

# Instantiation du modÃ¨le dont on veut optimiser les hyper-paramÃ¨tres.
model = AdaBoostRegressor(
    random_state=TRAIN_RANDOM_SEED
)

# Grille d'hyper-paramÃ¨tres Ã  tester
param_grid = {
    'base_estimator': [
        ExtraTreeRegressor(max_depth=29,criterion='mse',max_features=22,random_state=TRAIN_RANDOM_SEED),
        AdaBoostRegressor(
            base_estimator=ExtraTreeRegressor(
                max_depth=32,
                criterion='mse',
                max_features=22
            ),
            n_estimators=50,
            random_state=TRAIN_RANDOM_SEED
        )
    ]
}

# Instantiation du GridSearchCV.
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, n_jobs=-1, verbose=10)

# EntraÃ®nement du modÃ¨le
grid.fit(X_train, y_train)

# Affichage des meilleurs paramÃ¨tres
print(grid.best_params_)

# PrÃ©diction sur le corpus d'apprentissage afin de calculer l'erreur en prÃ©diction
y_train_predicted = grid.predict(X_train)
print("MeaAE en prÃ©diction : %i" % meaae(y_train, y_train_predicted))
print("MedAE en prÃ©diction : %i" % medae(y_train, y_train_predicted))
print("RMSE en prÃ©diction : %i" % rmse(y_train, y_train_predicted))

# PrÃ©diction sur le corpus de test afin de calculer l'erreur en gÃ©nÃ©ralisation
y_test_predicted = grid.predict(X_test)
print("MeaAE en gÃ©nÃ©ralisation : %i" % meaae(y_test, y_test_predicted))
print("MedAE en gÃ©nÃ©ralisation : %i" % medae(y_test, y_test_predicted))
print("RMSE en gÃ©nÃ©ralisation : %i" % rmse(y_test, y_test_predicted))
```

```python
s = 5

pe = ((y_test - y_test_predicted).abs() / y_test) * 100
rpe = np.floor(pe / s) * s

counts = rpe.value_counts()
counts = counts.reindex(index=np.sort(counts.index)).cumsum()

fig = go.Figure(data=go.Scatter(x=list(map(lambda x: "[0;%i[" % (x+s), counts.index)), y=(counts.values / len(rpe)) * 100, fill='tozeroy'))
fig.show()
```
