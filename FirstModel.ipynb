```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_pickle('data/checkpoints/after_analyze.pkl.xz')
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
df.drop(['longitude', 'latitude'], axis=1, inplace=True)
df.head()
```

```python
df.drop('surface_carrez', axis=1, inplace=True)
df.head()
```

```python
df['surface_terrain'].fillna(df['surface_reelle_bati'], inplace=True)
df.head()
```

```python
df['surface_non_bati'] = df['surface_terrain'] - df['surface_reelle_bati']
df.head()
```

```python
df['annee_mutation'] = df['date_mutation'].apply(lambda d: d.split('-')[0]).astype('int64')
df['mois_mutation'] = df['date_mutation'].apply(lambda d: d.split('-')[1]).astype('int64')
df.drop('date_mutation', axis=1, inplace=True)
df.head()
```

```python
#df['mois_mutation_sin'] = np.sin(2 * np.pi * (df['mois_mutation'] - 1) / 12)
#df['mois_mutation_cos'] = np.cos(2 * np.pi * (df['mois_mutation'] - 1) / 12)
#df.drop('mois_mutation', axis=1, inplace=True)
#df.head()
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
df.drop('type_local_Maison', axis=1, inplace=True)
df.head()
```

```python
df = df.T.drop_duplicates(keep='last').T
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Taille du dataset de test (fraction du dataset total)
TEST_SIZE = 0.30

# Random seed pour la sélection des observations
SPLIT_RANDOM_SEED = 42

# X et y
X = df.drop("valeur_fonciere", axis=1)
y = df['valeur_fonciere']

# Split du dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=SPLIT_RANDOM_SEED)
```

```python
print('Train dataset : X_train shape ' + str(X_train.shape) + ' et y_train shape ' + str(y_train.shape))
print('Test dataset : X_test shape ' + str(X_test.shape) + ' et y_test shape ' + str(y_test.shape))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, median_absolute_error

def meaae(y_true, y_pred):
    return mean_absolute_error(y_true, y_pred)

def medae(y_true, y_pred):
    return median_absolute_error(y_true, y_pred)

def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
from sklearn.ensemble import ExtraTreesRegressor

TRAIN_RANDOM_SEED=42

# Instantiation du modèle.
etr = ExtraTreesRegressor(n_estimators=25, criterion='mae', random_state=TRAIN_RANDOM_SEED, n_jobs=-1)

# Entraînement du modèle
etr.fit(X_train, y_train)

# Prédiction sur le corpus d'apprentissage afin de calculer l'erreur en prédiction
y_train_predicted = etr.predict(X_train)
print("MeaAE en prédiction : %i" % meaae(y_train, y_train_predicted))
print("MedAE en prédiction : %i" % medae(y_train, y_train_predicted))
print("RMSE en prédiction : %i" % rmse(y_train, y_train_predicted))

# Prédiction sur le corpus de test afin de calculer l'erreur en généralisation
y_test_predicted = etr.predict(X_test)
print("MeaAE en généralisation : %i" % meaae(y_test, y_test_predicted))
print("MedAE en généralisation : %i" % medae(y_test, y_test_predicted))
print("RMSE en généralisation : %i" % rmse(y_test, y_test_predicted))
```

```python
# Obtention de l'importance des features
importances = etr.feature_importances_

# Création d'un DataFrame pour associer chaque feature à son importance
importances = pd.concat([pd.Series(X_train.columns), pd.Series(importances)], axis=1).rename({0: 'feature', 1: 'importance'}, axis=1)

# Tri
importances.sort_values(by='importance', ascending=False, inplace=True)

#Affichage
importances
```

```python
 # Fusion de X_train et y_train pour obtenir le corpus d'entraînement en entier
train = X_train.copy()
train['valeur_fonciere'] = y_train

from sklearn.ensemble import IsolationForest

OUT_DET_RANDOM_STATE = 42

# Taux de contamination
CONTAMINATION=0.1

# Création de notre détecteur
outlier_detector = IsolationForest(n_estimators=500, contamination=CONTAMINATION, max_samples=1.0, behaviour='new', random_state=OUT_DET_RANDOM_STATE, n_jobs=-1)

# Analyse des données
outlier_detector.fit(train)

# Détermination des outliers
train['outlier'] = outlier_detector.predict(train)

print("L'Isolation Forest a trouvé %i outliers" % train[train['outlier'] < 0].shape[0])
```

```python
X_train2 = train[train['outlier'] > 0].drop(['valeur_fonciere', 'outlier'], axis=1)
y_train2 = train[train['outlier'] > 0]['valeur_fonciere']

print('Train dataset : X_train shape ' + str(X_train2.shape) + ' et y_train shape ' + str(y_train2.shape))
```

```python
from sklearn.ensemble import ExtraTreesRegressor

TRAIN_RANDOM_SEED=42

# Instantiation du modèle.
etr = ExtraTreesRegressor(n_estimators=25, criterion='mae', random_state=TRAIN_RANDOM_SEED, n_jobs=-1)

# Entraînement du modèle
etr.fit(X_train2, y_train2)

# Prédiction sur le corpus d'apprentissage afin de calculer l'erreur en prédiction
y_train_predicted = etr.predict(X_train2)
print("MeaAE en prédiction : %i" % meaae(y_train2, y_train_predicted))
print("MedAE en prédiction : %i" % medae(y_train2, y_train_predicted))
print("RMSE en prédiction : %i" % rmse(y_train2, y_train_predicted))

# Prédiction sur le corpus de test afin de calculer l'erreur en généralisation
y_test_predicted = etr.predict(X_test)
print("MeaAE en généralisation : %i" % meaae(y_test, y_test_predicted))
print("MedAE en généralisation : %i" % medae(y_test, y_test_predicted))
print("RMSE en généralisation : %i" % rmse(y_test, y_test_predicted))
```

```python
from sklearn.ensemble import ExtraTreesRegressor

TRAIN_RANDOM_SEED=42

# Instantiation du modèle.
etr = ExtraTreesRegressor(n_estimators=100, criterion='mae', random_state=TRAIN_RANDOM_SEED, n_jobs=-1, max_features='sqrt', bootstrap=True, oob_score=True)

# Entraînement du modèle
etr.fit(X_train2, y_train2)

# Prédiction sur le corpus d'apprentissage afin de calculer l'erreur en prédiction
y_train_predicted = etr.predict(X_train2)
print("MeaAE en prédiction : %i" % meaae(y_train2, y_train_predicted))
print("MedAE en prédiction : %i" % medae(y_train2, y_train_predicted))
print("RMSE en prédiction : %i" % rmse(y_train2, y_train_predicted))

# Prédiction sur le corpus de test afin de calculer l'erreur en généralisation
y_test_predicted = etr.predict(X_test)
print("MeaAE en généralisation : %i" % meaae(y_test, y_test_predicted))
print("MedAE en généralisation : %i" % medae(y_test, y_test_predicted))
print("RMSE en généralisation : %i" % rmse(y_test, y_test_predicted))
```

```python
# Obtention de l'importance des features
importances = etr.feature_importances_

# Création d'un DataFrame pour associer chaque feature à son importance
importances = pd.concat([pd.Series(X_train.columns), pd.Series(importances)], axis=1).rename({0: 'feature', 1: 'importance'}, axis=1)

# Tri
importances.sort_values(by='importance', ascending=False, inplace=True)

#Affichage
importances
```

```python
 from sklearn.model_selection import GridSearchCV

# Instantiation du modèle dont on veut optimiser les hyper-paramètres.
model = ExtraTreesRegressor(n_estimators=100, criterion='mae', random_state=TRAIN_RANDOM_SEED, n_jobs=-1, bootstrap=True, oob_score=True)

# Grille d'hyper-paramètres à tester.
param_grid = {
    'max_features': [2, 3, 4, 5],
    'max_depth': [30, 35, 40]
}

# Instantiation du GridSearchCV.
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=2, n_jobs=-1, verbose=10)

# Entraînement du modèle
grid.fit(X_train, y_train)

# Affichage des meilleurs paramètres
print(grid.best_params_)

# Prédiction sur le corpus d'apprentissage afin de calculer l'erreur en prédiction
y_train_predicted = grid.predict(X_train2)
print("MeaAE en prédiction : %i" % meaae(y_train2, y_train_predicted))
print("MedAE en prédiction : %i" % medae(y_train2, y_train_predicted))
print("RMSE en prédiction : %i" % rmse(y_train2, y_train_predicted))

# Prédiction sur le corpus de test afin de calculer l'erreur en généralisation
y_test_predicted = grid.predict(X_test)
print("MeaAE en généralisation : %i" % meaae(y_test, y_test_predicted))
print("MedAE en généralisation : %i" % medae(y_test, y_test_predicted))
print("RMSE en généralisation : %i" % rmse(y_test, y_test_predicted))
```

```python
# Obtention de l'importance des features
importances = grid.best_estimator_.feature_importances_

# Création d'un DataFrame pour associer chaque feature à son importance
importances = pd.concat([pd.Series(X_train.columns), pd.Series(importances)], axis=1).rename({0: 'feature', 1: 'importance'}, axis=1)

# Tri
importances.sort_values(by='importance', ascending=False, inplace=True)

#Affichage
importances
```
